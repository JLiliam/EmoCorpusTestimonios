###Script para guardar en una carpeta solo las transcripciones (PDF) de entrevistas asociadas con desplazamiento.
Script generado por ChatGPT

```python
import os
import shutil
import pandas as pd

# Ruta al archivo Excel
ruta_excel = 'C:\\Users\\Lenovo\\Desktop\\Tesis202420\\BasesDatos\\archivo-desplazamiento.xlsx'  # Cambia esto a la ruta de tu archivo Excel

# Ruta de la carpeta que contiene todos los archivos PDF
carpeta_pdf = 'C:\\Users\\Lenovo\\Desktop\\Tesis202420\\BasesDatos\\entrevistas_pdf\\'  # Cambia esto a la carpeta que contiene los PDF

# Ruta de la nueva carpeta donde se copiarán los archivos PDF seleccionados
nueva_carpeta = 'C:\\Users\\Lenovo\\Desktop\\Tesis202420\\BasesDatos\\entrevistas-desplazamiento\\'  # Cambia esto a donde quieres crear la nueva carpeta

# Lee el archivo Excel y obtiene la lista de archivos PDF
df = pd.read_excel(ruta_excel)

# Supongamos que los nombres de los archivos PDF están en la primera columna del Excel
lista_archivos = df.iloc[:, 0].tolist()

# Recorre la lista y copia los archivos PDF a la nueva carpeta
for archivo in lista_archivos:
    # Asegúrate de que cada nombre de archivo tiene la extensión .pdf
    if not archivo.endswith('.pdf'):
        archivo += '.pdf'
        
    archivo_pdf = os.path.join(carpeta_pdf, archivo)
    
    if os.path.isfile(archivo_pdf):
        shutil.copy(archivo_pdf, nueva_carpeta)
        print(f'Archivo copiado: {archivo}')
    else:
        print(f'Archivo no encontrado: {archivo}')

print('Proceso completado.')
```

```python
import os
import pdfplumber

# Ruta de la carpeta que contiene los archivos PDF
carpeta_pdf = 'C:\\Users\\Lenovo\\Desktop\\Tesis202420\\BasesDatos\\entrevistas-desplazamiento'

# Ruta de la carpeta donde se guardarán los archivos de texto
carpeta_texto = 'C:\\Users\\Lenovo\\Desktop\\Tesis202420\\BasesDatos\\entrevistas-desplazamiento\\archivos-desplazamiento-plano'

# Recorre cada archivo en la carpeta de archivos PDF
for archivo in os.listdir(carpeta_pdf):
    # Comprobar si el archivo es un PDF
    if archivo.endswith('.pdf'):
        # Ruta completa al archivo PDF
        ruta_pdf = os.path.join(carpeta_pdf, archivo)
        
        # Ruta completa del archivo de texto donde se guardará el contenido
        nombre_archivo_txt = os.path.splitext(archivo)[0] + '.txt'
        ruta_texto = os.path.join(carpeta_texto, nombre_archivo_txt)
        
        # Abrir el PDF y extraer el texto
        with pdfplumber.open(ruta_pdf) as pdf:
            texto = ''
            for pagina in pdf.pages:
                texto += pagina.extract_text()
        
        # Guardar el texto extraído en un archivo de texto
        with open(ruta_texto, 'w', encoding='utf-8') as archivo_txt:
            archivo_txt.write(texto)
        
        print(f'Texto extraído y guardado en: {ruta_texto}')

print('Proceso completado.')
```

###Obtención de cantidad de palabras de cada entrevista relacionada con desplazamiento
Script generado por ChatGPT

```python
import os
import pandas as pd

# Especificar la ruta de la carpeta con los archivos de texto
ruta_carpeta = 'C:\\Users\\Lenovo\\Desktop\\Tesis202420\\BasesDatos\\entrevistas-desplazamiento\\archivos-desplazamiento-plano'

# Lista para almacenar los resultados
resultados = []

# Recorrer cada archivo en la carpeta
for nombre_archivo in os.listdir(ruta_carpeta):
    # Verificar que el archivo sea de texto
    if nombre_archivo.endswith('.txt'):
        # Ruta completa del archivo
        ruta_archivo = os.path.join(ruta_carpeta, nombre_archivo)
        
        # Contar palabras en el archivo
        with open(ruta_archivo, 'r', encoding='utf-8') as archivo:
            contenido = archivo.read()
            cantidad_palabras = len(contenido.split())
        
        # Agregar el resultado a la lista
        resultados.append({'Nombre del archivo': nombre_archivo, 'Cantidad de palabras': cantidad_palabras})

# Crear un DataFrame de Pandas con los resultados
df_resultados = pd.DataFrame(resultados)

# Mostrar el DataFrame
df_resultados
```

```python
# Guardar el DataFrame en un archivo CSV
df_resultados.to_csv('conteo_palabras_entrevistas_desplazamiento.csv', index=False)
```
